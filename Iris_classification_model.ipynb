```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from IPython.display import Image #Image용


```

라이브러리 임포트

pandas as pd: 데이터 처리용

load_iris: 사이킷런의 붓꽃(Iris) 데이터셋 로드

train_test_split: 데이터셋을 훈련/테스트 세트로 분할

RandomForestClassifier: 랜덤 포레스트 분류기 모델

accuracy_score: 모델 정확도 평가 지표

데이터 로드

iris = load_iris(): 붓꽃 데이터셋 로드 (150개의 샘플, 4개의 특징, 3개 클래스)

데이터 분할

train_test_split을 이용해 데이터 분리

test_size=0.2: 테스트 데이터 20%

random_state=42: 결과 재현성을 위한 시드 설정

결과: X_train, X_test, y_train, y_test

모델 생성 및 학습

clf = RandomForestClassifier(): 랜덤 포레스트 모델 생성

clf.fit(X_train, y_train): 훈련 데이터로 모델 학습

예측 수행

y_pred = clf.predict(X_test): 테스트 데이터로 예측 수행

정확도 평가

accuracy_score(y_test, y_pred): 예측 결과와 실제 값을 비교해 정확도 계산

print("Accuracy:", ...): 정확도 출력

🪻🪻 데이터 소개 (붓꽃 - Iris)
품종 3가지: Setosa(세토사), Versicolor(버시컬러), Virginica(버지니카)

각 꽃은 다음 4가지 특성
1. 꽃받침 길이(sepal length)
2. 꽃받침 너비(sepal width)
3. 꽃잎 길이(petal length)
4. 꽃잎 너비(petal width)

```python
#Step1. 데이터 불러오기
iris = load_iris()
X = iris.data
Y = iris.target
# print(X)
# print('------------------------')
# print(Y)
# print(len(X),len(Y))

#데이터 display
pd.DataFrame(X, columns = iris.feature_names).head()
```

```python
#Step 2:AI가 공부할 데이터 준비
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 42)
```

```python
#Step 3: AI 모델 만들기
model = RandomForestClassifier()
model.fit(X_train, Y_train)

```

```python
#Step 4: AI가 test를 보게 하기
Y_pred = model.predict(X_test)
acc = accuracy_score(Y_test, Y_pred)
print('Accuracy:', acc*100, '%')
```
